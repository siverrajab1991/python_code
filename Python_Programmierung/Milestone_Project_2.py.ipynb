```python
import random 
```

```python
Suits = ("Herz","Diamant","Spades","Clubs")
Karten = ("Zwei","Drei","Vier","Fünf","Sechs","Sieben","Acht","Neun","Zehn","Bube","Dame","König","Ass")
Value = {"Zwei":2,"Drei":3,"Vier":4,"Fünf":5,"Sechs":6,"Sieben":7,"Acht":8,"Neun":9,"Zehn":10,"Bube":10,"Dame":10,"König":10,"Ass":11}




class Card():
    
    def __init__(self,Suits,Karten):
        
        self.Suits = Suits
        self.Karten = Karten 
        self.Value = Value[Karten]
        
        
    
    def __str__(self):
        return self.Suits + " von " + self.Karten
    
    
    def __len__(self):
        return self.value
    
    

    
```

```python
class Deck():
    
    
    def __init__(self):
        
        self.cards = []
        for suit in Suits:
            for kart in Karten:
                self.cards.append(Card(suit,kart))
                
    
    def shuffle(self):
        random.shuffle(self.cards)
        
    
    def remove_card(self):
        
        if self.cards:
            self.cards.pop(0)
            
            
    def cards_left(self):
        
        if len(self.cards) < 4:
            return False
            print("Es sind nicht mehr genügend Karten übrig für eine neue Runde")
        else:
            return True
            
```

```python
class Bet():
    
    def __init__(self,bet):
        
        self.bet = bet
        bet = 0
        

    def place_bet(self):
        while True:
            try:
                a = abs(int(input("Was ist dein Einsatz? ")))
            except ValueError: 
                print("Ungültige Eingabe! Bitte einen Integer-Wert eingeben.")
                continue

            if a > self.bet:
                print(f"Das ist mehr als du übrig hast, denn du hast nur noch {self.bet}")
            else:
                self.bet -= a
                print(f"Dein Einsatz beträgt {a} und du hast noch {self.bet} übrig")
                return a


    
    def get_win(self,w):
        
        self.bet += (w*2)
        print(f"Dank dem Gewinn hast du nun {self.bet}")
        return self.bet
        
    
    def kontostand(self):
        
        return self.bet
    
    
        
        
    
    
```
